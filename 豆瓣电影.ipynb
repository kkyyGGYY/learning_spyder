```python
import requests
import re
```

```python
content = requests.get('https://movie.douban.com/').text
print(content)
```

```python
partten = re.compile('<li.*?title="(.*?)".*?ease="(.*?)".*?trailer="(.*?)".*?duration="(.*?)".*?director="(.*?)".*?actors="(.*?)".*?</li>', re.S)
results = re.findall(partten,content)
for result in results:
    name, year, url, time, director, actor = result
    print(name, year, url, time, director, actor.split(' / '))
# print(results)
```
